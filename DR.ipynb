```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import umap.umap_ as umap  # Requires 'pip install umap-learn'

# We generate synthetic high-dimensional data (e.g., 1000 samples, 20 features),
# you can play around with this
X, y = make_blobs(n_samples=1000, n_features=20, centers=5, random_state=42)

# Apply PCA
# Note: PCA is good with a linear technique
# Global
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Apply t-SNE
# non-linear, focuses on local
tsne = TSNE(n_components=2, random_state=42, init="random", learning_rate=200)
X_tsne = tsne.fit_transform(X)

# Apply UMAP
# non-linear, balance of both local and some global structure
reducer = umap.UMAP(n_components=2, random_state=42)
X_umap = reducer.fit_transform(X)

# 5. Plot each 2D projection (PCA, t-SNE, UMAP)

plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title("Original Data (using Features 0 & 1)")
plt.show()

# PCA visualization
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y)
plt.title("PCA 2D Projection")
plt.show()

# t-SNE visualization
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y)
plt.title("t-SNE 2D Projection")
plt.show()

# UMAP visualization
plt.scatter(X_umap[:, 0], X_umap[:, 1], c=y)
plt.title("UMAP 2D Projection")
plt.show()
```
