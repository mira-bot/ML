```python
import numpy as np
import matplotlib.pyplot as plt

# A non-convex function in 2D: f(x, y) = x^2 + y^2 + 5 * sin(x) * sin(y)
def f(x, y):
    return x**2 + y**2 + 5.0 * np.sin(x) * np.sin(y)

# Generate a grid
x_vals = np.linspace(-4, 4, 200)
y_vals = np.linspace(-4, 4, 200)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f(X, Y)

# Plot contour
plt.figure(figsize=(6,5))
cs = plt.contour(X, Y, Z, levels=30, cmap='jet')
plt.clabel(cs, inline=1, fontsize=8)
plt.title("A Non-Convex Function with Multiple Minima")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
